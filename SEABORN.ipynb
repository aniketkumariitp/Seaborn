#$$Distributionplot$$

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### üìä Distribution Plots in Seaborn

We use **distribution plots** to visualize the **distribution of quantitative data**.  

---

#### üîé Common distribution plots:
- **`distplot` / `histplot`** ‚Üí Histogram with optional KDE
- **`jointplot`** ‚Üí Relationship between two variables with distribution
- **`pairplot`** ‚Üí Pairwise relationships in a dataset
- **`rugplot`** ‚Üí Small ticks along an axis to show data distribution


## 1Ô∏è‚É£ distplot / histplot
- **Use case**: Histogram to visualize the distribution of a single variable.  
- **Extra**: Can add **KDE (Kernel Density Estimate)** for smooth curve.  


```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
```

```python
sns.pairplot(df,hue = 'sex')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df['tip'])
```

#$$Categoricalplots$$

# üìä Categorical Data Plots

We use **categorical plots** in Seaborn to visualize relationships between **categorical variables and numerical values**.  

---

#### üîé Common categorical plots:
- **`boxplot`** ‚Üí Shows distribution using median, quartiles, and outliers  
- **`violinplot`** ‚Üí Combines boxplot with a KDE (shows density)  
- **`stripplot`** ‚Üí Scatter plot for categorical data  
- **`swarmplot`** ‚Üí Like stripplot but with non-overlapping points  
- **`barplot`** ‚Üí Shows average values with confidence intervals  
- **`countplot`** ‚Üí Shows count of each category  


## 1Ô∏è‚É£ boxplot
- **Use case**: Shows distribution of data using **median, quartiles, and outliers**.  
- **Extra**: Useful for comparing **multiple categories** side by side.  


```python
df = sns.load_dataset("tips")
df
```

```python
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
sns.barplot(x = df['sex'],y = df['total_bill'])
```

```python
import numpy as np
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
```

```python
sns.boxplot(x= 'tip',y = 'day',data = df,palette = 'rainbow')
```

## 2Ô∏è‚É£ violinplot
- **Use case**: Combines a **boxplot** with a **KDE curve** to show distribution density.  
- **Extra**: Helps in understanding the **shape of distribution**.  


```python
sns.violinplot(x = 'day',y = 'total_bill',data = df)
```

## 3Ô∏è‚É£ stripplot
- **Use case**: Displays **individual data points** for categorical variables.  
- **Extra**: Points may **overlap** when dataset is large.  


```python
sns.stripplot(x = 'day',y = 'total_bill',data = df)
```

## 4Ô∏è‚É£ swarmplot
- **Use case**: Similar to stripplot but avoids overlapping points.  
- **Extra**: Best for **small datasets** to clearly see all observations.  


```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df)
```

# üß© **Matrix Plots in Seaborn**

Matrix plots are used to visualize **data in a tabular or grid form** with colors representing values.  
The main types are: **Heatmap, Clustermap, and Pivot Table**.  

---

## 1Ô∏è‚É£ Heatmap
- **Use case**: Displays data values as a **color-encoded matrix**.  
- **Extra**: Often used with **correlation matrices** or pivot tables.  
- **Tip**: Can add **annotations** to display values inside cells.  

---

## 2Ô∏è‚É£ Clustermap
- **Use case**: Similar to heatmap but performs **hierarchical clustering** on rows and columns.  
- **Extra**: Groups similar data together for better pattern detection.  
- **Tip**: Useful for gene expression data, similarity matrices, etc.  

---

## 3Ô∏è‚É£ Pivot Table
- **Use case**: Creates a **summarized 2D matrix** (rows √ó columns) from raw data.  
- **Extra**: Often used as input for a seaborn **heatmap**.  
- **Tip**: Created using `pandas.pivot_table()` before visualization.  


## 1Ô∏è‚É£ Heatmap
- **Use case**: Displays data values as a **color-encoded matrix**.  
- **Extra**: Often used with **correlation matrices** or pivot tables.  
- **Tip**: Can add **annotations** to display values inside cells.  

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tipscorr  = tips[['total_bill','tip','size']]
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot= True)
```

## 2Ô∏è‚É£ Clustermap
- **Use case**: Similar to heatmap but performs **hierarchical clustering** on rows and columns.  
- **Extra**: Groups similar data together for better pattern detection.  
- **Tip**: Useful for gene expression data, similarity matrices, etc.  


```python
sns.clustermap(tipscorr.corr())
```

## 3Ô∏è‚É£ Pivot Table
- **Use case**: Creates a **summarized 2D matrix** (rows √ó columns) from data.  
- **Extra**: Often used as the input for a seaborn **heatmap**.  
- **Tip**: Made using `pandas.pivot_table()` before passing into seaborn plots.  


```python
pvtflight = flights.pivot_table(values = 'passengers',index = 'month',columns = 'year')
```

```python
pvtflight
```

```python
sns.heatmap(pvtflight)
```

# üìà **Regression Plots in Seaborn**

Regression plots are used to **visualize relationships between two variables** and fit **regression models** (like straight lines).  
The main types are: **regplot** and **lmplot**.  



## 1 lmplot
- **Use case**: High-level function for regression plots.  
- **Extra**: Can create **multiple regression plots** by grouping data using `hue`, `col`, or `row`.  
- **Tip**: Built on top of `regplot`, but better for **faceting and categorical separation**.  


```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':50})
```
